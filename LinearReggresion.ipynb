```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Simulate example data
np.random.seed(42)
market = np.random.normal(0.001, 0.02, 100)          # market returns
stock  = 0.5 * market + np.random.normal(0, 0.01, 100)  # stock ~ market + noise
df = pd.DataFrame({'Market': market, 'Stock': stock})

# Prepare regression inputs
X = df[['Market']]   # independent variable(s)
y = df['Stock']      # dependent variable

# Fit linear regression (OLS)
model = LinearRegression().fit(X, y)
print("Intercept (alpha):", model.intercept_)
print("Beta (slope):", model.coef_[0])
```

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X)            # add constant term for intercept
ols_model = sm.OLS(y, X_sm).fit()
print(ols_model.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset
# We'll create a linear relationship: y = 2*x + 10 + error
n_observations = 100
x = np.linspace(1, 100, n_observations)
error = np.random.normal(0, 50, n_observations) # Normal random error
y = 2 * x + 10 + error

# Create a DataFrame for easier handling
df = pd.DataFrame({'x1': x, 'y': y})

print("Generated a dataset with", n_observations, "observations.")
```

```python
# Add a constant (intercept) to the independent variable
X = sm.add_constant(df['x1'])
Y = df['y']

# Fit the OLS model
model = sm.OLS(Y, X)
results = model.fit()

# Print the summary
print(results.summary())
```

```python
# Create a scatter plot of the data with the regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='x1', y='y', data=df, ci=95)
plt.title('Scatter Plot with OLS Regression Line')
plt.xlabel('x1')
plt.ylabel('y')
plt.show()

# Visualize Residuals: Checking for Normality
# Plot 1: Histogram of residuals
residuals = results.resid
plt.figure(figsize=(10, 5))
sns.histplot(residuals, kde=True)
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.show()

# Plot 2: Q-Q Plot of residuals
sm.qqplot(residuals, line='45', fit=True)
plt.title('Q-Q Plot of Residuals')
plt.show()

# Visualize Residuals: Checking for Homoscedasticity
# Plot 3: Residuals vs Fitted Values
plt.figure(figsize=(10, 6))
plt.scatter(results.fittedvalues, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, normaltest, jarque_bera
from statsmodels.stats.stattools import durbin_watson

# Set style for better plots
plt.style.use('default')
sns.set_palette("husl")

print("=" * 60)
print("RESIDUAL DIAGNOSTICS ANALYSIS")
print("=" * 60)

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with a linear relationship
n_observations = 200  # A larger sample size for better test reliability
x = np.linspace(1, 200, n_observations)
error = np.random.normal(0, 20, n_observations)  # Normally distributed random error
y = 3 * x + 50 + error

# Fit the OLS model
X = sm.add_constant(x)
model = sm.OLS(y, X)
results = model.fit()

# Get the residuals from the fitted model
residuals = results.resid

print(f"Dataset generated with {n_observations} observations")
print(f"Model R-squared: {results.rsquared:.4f}")
print("\n" + "=" * 60)

# =============================================================================
# 1. OMNIBUS TEST AND JARQUE-BERA TEST
# =============================================================================
print("1. OMNIBUS AND JARQUE-BERA TESTS (Normality)")
print("-" * 50)

# Calculate the Omnibus and Jarque-Bera statistics
omnibus_stat, omnibus_p_value = normaltest(residuals)
jb_stat, jb_p_value = jarque_bera(residuals)

print(f"Omnibus Test Statistic: {omnibus_stat:.4f}, p-value: {omnibus_p_value:.4f}")
print(f"Jarque-Bera Test Statistic: {jb_stat:.4f}, p-value: {jb_p_value:.4f}")

# Interpretation
if omnibus_p_value > 0.05:
    print(" Omnibus Test: Residuals appear normally distributed (p > 0.05)")
else:
    print(" Omnibus Test: Residuals may not be normally distributed (p  0.05)")

if jb_p_value > 0.05:
    print(" Jarque-Bera Test: Residuals appear normally distributed (p > 0.05)")
else:
    print(" Jarque-Bera Test: Residuals may not be normally distributed (p  0.05)")

# Create a figure with two subplots for visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Histogram of residuals
sns.histplot(residuals, kde=True, ax=ax1, stat='density')
ax1.set_title('Histogram of Residuals (Distribution Check)')
ax1.set_xlabel('Residuals')
ax1.axvline(np.mean(residuals), color='red', linestyle='--', label=f'Mean: {np.mean(residuals):.2f}')
ax1.legend()

# Q-Q Plot (Quantile-Quantile Plot)
sm.qqplot(residuals, line='45', ax=ax2)
ax2.set_title('Q-Q Plot of Residuals (Normality Check)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nVisual Interpretation:")
print(" Histogram: A bell-shaped curve indicates normality")
print(" Q-Q Plot: Points close to the red 45-degree line indicate normal distribution")
print("\n" + "=" * 60)

# =============================================================================
# 2. SKEWNESS
# =============================================================================
print("2. SKEWNESS ANALYSIS")
print("-" * 50)

# Calculate Skewness
calculated_skewness = skew(residuals)
print(f"Calculated Skewness: {calculated_skewness:.4f}")

# Interpretation
if abs(calculated_skewness) < 0.5:
    print(" Skewness: Distribution is approximately symmetric")
elif abs(calculated_skewness) < 1:
    print(" Skewness: Distribution is moderately skewed")
else:
    print(" Skewness: Distribution is highly skewed")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Histogram with a skewed distribution for comparison
np.random.seed(123)  # Different seed for comparison
skewed_data = np.random.gamma(2, 2, 1000)
sns.histplot(skewed_data, kde=True, ax=ax1, stat='density')
ax1.set_title(f'Example of Positive Skewness\n(Skew: {skew(skewed_data):.2f})')
ax1.set_xlabel('Value')

# Histogram of our residuals
sns.histplot(residuals, kde=True, ax=ax2, stat='density')
ax2.set_title(f'OLS Residuals\n(Skew: {calculated_skewness:.2f})')
ax2.set_xlabel('Residuals')
ax2.axvline(np.mean(residuals), color='red', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

print("\nVisual Interpretation:")
print(" Skewness  0: Symmetric distribution (ideal)")
print(" Skewness > 0: Right tail is longer (positive skew)")
print(" Skewness < 0: Left tail is longer (negative skew)")
print("\n" + "=" * 60)

# =============================================================================
# 3. KURTOSIS
# =============================================================================
print("3. KURTOSIS ANALYSIS")
print("-" * 50)

# Calculate Kurtosis
calculated_kurtosis = kurtosis(residuals, fisher=True)  # fisher=True gives excess kurtosis (Normal = 0)
print(f"Calculated Excess Kurtosis: {calculated_kurtosis:.4f}")

# Interpretation
if abs(calculated_kurtosis) < 0.5:
    print(" Kurtosis: Distribution has normal tail behavior")
elif abs(calculated_kurtosis) < 1:
    print(" Kurtosis: Distribution has moderate tail deviation from normal")
else:
    if calculated_kurtosis > 0:
        print(" Kurtosis: Distribution has heavy tails (leptokurtic)")
    else:
        print(" Kurtosis: Distribution has light tails (platykurtic)")

# Create a figure with two subplots for visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Example of a Leptokurtic distribution
np.random.seed(456)
leptokurtic_data = np.random.standard_t(4, 1000)
sns.histplot(leptokurtic_data, kde=True, ax=ax1, stat='density')
ax1.set_title(f'Example: Leptokurtic Distribution\n(Excess Kurtosis: {kurtosis(leptokurtic_data):.2f})')
ax1.set_xlabel('Value')

# Histogram of our residuals
sns.histplot(residuals, kde=True, ax=ax2, stat='density')
ax2.set_title(f'OLS Residuals\n(Excess Kurtosis: {calculated_kurtosis:.2f})')
ax2.set_xlabel('Residuals')

plt.tight_layout()
plt.show()

print("\nVisual Interpretation:")
print(" Excess Kurtosis  0: Normal tail behavior (ideal)")
print(" Excess Kurtosis > 0: Heavy tails, sharp peak (leptokurtic)")
print(" Excess Kurtosis < 0: Light tails, flat peak (platykurtic)")
print("\n" + "=" * 60)

# =============================================================================
# 4. DURBIN-WATSON TEST
# =============================================================================
print("4. DURBIN-WATSON TEST (Autocorrelation)")
print("-" * 50)

# Calculate the Durbin-Watson statistic
dw_statistic = durbin_watson(residuals)
print(f"Durbin-Watson Statistic: {dw_statistic:.4f}")

# Interpretation
if 1.5 <= dw_statistic <= 2.5:
    print(" Durbin-Watson: No significant autocorrelation detected")
elif dw_statistic < 1.5:
    print(" Durbin-Watson: Positive autocorrelation detected")
else:
    print(" Durbin-Watson: Negative autocorrelation detected")

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Residuals over time (or observation index)
ax1.plot(residuals, alpha=0.7)
ax1.axhline(y=0, color='r', linestyle='--', alpha=0.8)
ax1.set_title('Residuals Over Time')
ax1.set_xlabel('Observation Index')
ax1.set_ylabel('Residuals')
ax1.grid(True, alpha=0.3)

# Plot 2: Lagged residuals
lag_residuals = pd.Series(residuals).shift(1).dropna()
current_residuals = residuals[1:]
ax2.scatter(x=lag_residuals, y=current_residuals, alpha=0.6)
ax2.set_title('Residuals vs. Lagged Residuals')
ax2.set_xlabel('Residuals at t-1')
ax2.set_ylabel('Residuals at t')
ax2.axhline(y=0, color='r', linestyle='--', alpha=0.8)
ax2.axvline(x=0, color='r', linestyle='--', alpha=0.8)
ax2.grid(True, alpha=0.3)

# Add correlation coefficient
correlation = np.corrcoef(lag_residuals, current_residuals)[0, 1]
ax2.text(0.05, 0.95, f'Correlation: {correlation:.3f}', 
         transform=ax2.transAxes, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

print("\nVisual Interpretation:")
print(" Time Series Plot: Random scatter around zero indicates no autocorrelation")
print(" Lag Plot: Random cloud of points indicates no autocorrelation")
print(" Pattern in lag plot suggests autocorrelation is present")
print("\n" + "=" * 60)

# =============================================================================
# SUMMARY REPORT
# =============================================================================
print("SUMMARY REPORT")
print("-" * 50)
print(f"Sample Size: {n_observations}")
print(f"Model R-squared: {results.rsquared:.4f}")
print()
print("Diagnostic Test Results:")
print(f" Omnibus Test p-value: {omnibus_p_value:.4f} {'' if omnibus_p_value > 0.05 else ''}")
print(f" Jarque-Bera Test p-value: {jb_p_value:.4f} {'' if jb_p_value > 0.05 else ''}")
print(f" Skewness: {calculated_skewness:.4f} {'' if abs(calculated_skewness) < 0.5 else '' if abs(calculated_skewness) < 1 else ''}")
print(f" Excess Kurtosis: {calculated_kurtosis:.4f} {'' if abs(calculated_kurtosis) < 0.5 else '' if abs(calculated_kurtosis) < 1 else ''}")
print(f" Durbin-Watson: {dw_statistic:.4f} {'' if 1.5 <= dw_statistic <= 2.5 else ''}")
print()
print("Legend:  = Good,  = Acceptable,  = Concerning")
print("=" * 60)
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import het_white
from statsmodels.stats.stattools import durbin_watson

# Set random seed for reproducibility
np.random.seed(42)

print("=" * 70)
print("MODEL STATISTICS DEMONSTRATION")
print("=" * 70)

# Generate synthetic dataset
n = 100  # sample size
x1 = np.random.normal(0, 1, n)
x2 = np.random.normal(0, 1, n)
x3 = np.random.normal(0, 1, n)
error = np.random.normal(0, 2, n)

# True relationship
y = 2 + 3*x1 + 1.5*x2 + 0.5*x3 + error

# Create DataFrame
data = pd.DataFrame({
    'y': y,
    'x1': x1, 
    'x2': x2,
    'x3': x3
})

print(f"Dataset created with {n} observations and 3 predictors")
print("\n" + "=" * 70)

# =============================================================================
# MODEL COMPARISON: Different complexities
# =============================================================================
print("MODEL COMPARISON: Different Complexities")
print("-" * 50)

# Model 1: Simple (only x1)
X1 = sm.add_constant(data[['x1']])
model1 = sm.OLS(data['y'], X1).fit()

# Model 2: Medium (x1, x2)
X2 = sm.add_constant(data[['x1', 'x2']])
model2 = sm.OLS(data['y'], X2).fit()

# Model 3: Full (x1, x2, x3)
X3 = sm.add_constant(data[['x1', 'x2', 'x3']])
model3 = sm.OLS(data['y'], X3).fit()

# Model 4: Overfitted (x1, x2, x3, x1^2, x2^2)
data_expanded = data.copy()
data_expanded['x1_sq'] = data['x1']**2
data_expanded['x2_sq'] = data['x2']**2
X4 = sm.add_constant(data_expanded[['x1', 'x2', 'x3', 'x1_sq', 'x2_sq']])
model4 = sm.OLS(data['y'], X4).fit()

# Create comparison table
models = [model1, model2, model3, model4]
model_names = ['Simple (x1)', 'Medium (x1,x2)', 'Full (x1,x2,x3)', 'Overfitted']

print(f"{'Model':<15} {'Log-Like':<10} {'Df Resid':<10} {'Df Model':<10} {'AIC':<10} {'BIC':<10}")
print("-" * 70)

for i, (model, name) in enumerate(zip(models, model_names)):
    print(f"{name:<15} {model.llf:<10.2f} {model.df_resid:<10.0f} {model.df_model:<10.0f} {model.aic:<10.2f} {model.bic:<10.2f}")

print("\n" + "=" * 70)

# =============================================================================
# DETAILED ANALYSIS OF BEST MODEL
# =============================================================================
print("DETAILED ANALYSIS: Full Model")
print("-" * 50)

best_model = model3  # Full model is theoretically correct

print("MODEL STATISTICS:")
print(f"Log-Likelihood: {best_model.llf:.4f}")
print(f"   Higher is better (measures how well model fits data)")
print(f"   This value depends on sample size and data scale")

print(f"\nDegrees of Freedom:")
print(f"  Df Residuals: {best_model.df_resid:.0f}")
print(f"   Formula: n - k = {n} - {len(best_model.params)} = {best_model.df_resid:.0f}")
print(f"   Used for statistical inference (t-tests, confidence intervals)")
print(f"  Df Model: {best_model.df_model:.0f}")
print(f"   Number of predictors (excluding intercept)")
print(f"   Used in F-statistic calculation")

print(f"\nModel Selection Criteria:")
print(f"  AIC: {best_model.aic:.4f}")
print(f"   Formula: 2k - 2Log-Likelihood = 2{len(best_model.params)} - 2{best_model.llf:.2f}")
print(f"   Lower is better, balances fit vs complexity")
print(f"  BIC: {best_model.bic:.4f}")
print(f"   Formula: kln(n) - 2Log-Likelihood")
print(f"   More conservative than AIC (stronger complexity penalty)")

print(f"\nCovariance Type: {best_model.cov_type}")
print(f"   'nonrobust' assumes homoscedasticity and no autocorrelation")
print(f"   Use 'HC1', 'HC2', 'HC3' for heteroscedasticity")
print(f"   Use 'HAC' for autocorrelation in time series")

print("\n" + "=" * 70)

# =============================================================================
# COVARIANCE TYPES DEMONSTRATION
# =============================================================================
print("COVARIANCE TYPES COMPARISON")
print("-" * 50)

# Fit same model with different covariance types
cov_types = ['nonrobust', 'HC1', 'HC2', 'HC3']
results_cov = {}

for cov_type in cov_types:
    if cov_type == 'nonrobust':
        results_cov[cov_type] = sm.OLS(data['y'], X3).fit()
    else:
        results_cov[cov_type] = sm.OLS(data['y'], X3).fit(cov_type=cov_type)

# Compare standard errors
print("Standard Errors Comparison:")
print(f"{'Parameter':<12} {'nonrobust':<12} {'HC1':<12} {'HC2':<12} {'HC3':<12}")
print("-" * 65)

for i, param in enumerate(['const', 'x1', 'x2', 'x3']):
    row = f"{param:<12}"
    for cov_type in cov_types:
        se = results_cov[cov_type].bse[i]
        row += f"{se:<12.4f}"
    print(row)

print("\nInterpretation:")
print(" Similar standard errors suggest homoscedasticity assumption is reasonable")
print(" Large differences would suggest heteroscedasticity")

# Test for heteroscedasticity
white_test_results = het_white(best_model.resid, best_model.model.exog)
white_stat, white_p = white_test_results[0], white_test_results[1]
print(f"\nWhite's Test for Heteroscedasticity:")
print(f"  Statistic: {white_stat:.4f}, p-value: {white_p:.4f}")
if white_p > 0.05:
    print("   No evidence of heteroscedasticity (nonrobust covariance OK)")
else:
    print("   Evidence of heteroscedasticity (consider HC1, HC2, or HC3)")

print("\n" + "=" * 70)

# =============================================================================
# MODEL SELECTION VISUALIZATION
# =============================================================================
print("MODEL SELECTION VISUALIZATION")
print("-" * 50)

# Create visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Log-Likelihood
log_likes = [model.llf for model in models]
ax1.bar(range(len(models)), log_likes, color='skyblue', alpha=0.7)
ax1.set_title('Log-Likelihood (Higher is Better)')
ax1.set_xlabel('Model Complexity')
ax1.set_ylabel('Log-Likelihood')
ax1.set_xticks(range(len(models)))
ax1.set_xticklabels(['Simple', 'Medium', 'Full', 'Overfitted'], rotation=45)
ax1.grid(True, alpha=0.3)

# Plot 2: AIC
aics = [model.aic for model in models]
ax2.bar(range(len(models)), aics, color='lightcoral', alpha=0.7)
ax2.set_title('AIC (Lower is Better)')
ax2.set_xlabel('Model Complexity')
ax2.set_ylabel('AIC')
ax2.set_xticks(range(len(models)))
ax2.set_xticklabels(['Simple', 'Medium', 'Full', 'Overfitted'], rotation=45)
ax2.grid(True, alpha=0.3)

# Plot 3: BIC
bics = [model.bic for model in models]
ax3.bar(range(len(models)), bics, color='lightgreen', alpha=0.7)
ax3.set_title('BIC (Lower is Better)')
ax3.set_xlabel('Model Complexity')
ax3.set_ylabel('BIC')
ax3.set_xticks(range(len(models)))
ax3.set_xticklabels(['Simple', 'Medium', 'Full', 'Overfitted'], rotation=45)
ax3.grid(True, alpha=0.3)

# Plot 4: AIC vs BIC comparison
ax4.plot(range(len(models)), aics, 'o-', label='AIC', linewidth=2, markersize=8)
ax4.plot(range(len(models)), bics, 's-', label='BIC', linewidth=2, markersize=8)
ax4.set_title('AIC vs BIC Comparison')
ax4.set_xlabel('Model Complexity')
ax4.set_ylabel('Information Criterion')
ax4.set_xticks(range(len(models)))
ax4.set_xticklabels(['Simple', 'Medium', 'Full', 'Overfitted'], rotation=45)
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Find best models
best_aic_idx = np.argmin(aics)
best_bic_idx = np.argmin(bics)

print(f"Best model by AIC: {model_names[best_aic_idx]} (AIC = {aics[best_aic_idx]:.2f})")
print(f"Best model by BIC: {model_names[best_bic_idx]} (BIC = {bics[best_bic_idx]:.2f})")

if best_aic_idx == best_bic_idx:
    print(" Both criteria agree on the best model")
else:
    print(" AIC and BIC suggest different models")
    print("   AIC tends to select more complex models")
    print("   BIC is more conservative (prefers simpler models)")

print("\n" + "=" * 70)

# =============================================================================
# SUMMARY AND RECOMMENDATIONS
# =============================================================================
print("SUMMARY AND RECOMMENDATIONS")
print("-" * 50)

print("Key Takeaways:")
print("1. Log-Likelihood: Measures model fit (higher is better)")
print("2. Df Residuals: Available for inference (n - k)")
print("3. Df Model: Model complexity (number of predictors)")
print("4. AIC/BIC: Balance fit vs complexity (lower is better)")
print("5. Covariance Type: Choose based on data assumptions")

print(f"\nFor this dataset:")
print(f" Sample size: {n} observations")
print(f" Best model has {best_model.df_resid:.0f} residual degrees of freedom")
print(f" {'Adequate' if best_model.df_resid >= 30 else 'Limited'} sample size for reliable inference")
print(f" Covariance assumption appears {'reasonable' if white_p > 0.05 else 'questionable'}")

print("\nModel Selection Guidelines:")
print(" Use AIC for prediction-focused applications")
print(" Use BIC for explanation-focused applications")
print(" Consider both, but BIC is more reliable with large samples")
print(" Always check residual assumptions")

print("=" * 70)
```
